#### 1. Cálculo del Índice de Madurez (IM) 🍎

Este código ayuda a determinar si un fruto está listo para la cosecha basándose en la relación entre sólidos solubles ($Brix$) y la acidez titulable.

#### 2. Control de Humedad en Granos 🌾

Un script sencillo que simula una alerta si la humedad de un lote de granos está fuera del rango seguro para almacenamiento.

```
humedad_actual = 15.5  # Porcentaje de humedad medido por un sensor

if humedad_actual > 14.0:
    print("ALERTA: Humedad alta. Activar ventiladores de secado.")
else:
    print("Estado óptimo: La humedad es adecuada para almacenamiento.")
```

#### 3. Conversión de Unidades de Temperatura 🌡️

En la industria alimentaria, es común trabajar con diferentes escalas. Este código convierte grados Celsius a Fahrenheit para un proceso de pasteurización.

```
temp_celsius = 72  # Temperatura de pasteurización en °C

# Fórmula de conversión
temp_fahrenheit = (temp_celsius * 9/5) + 32

print(f"La temperatura de {temp_celsius}°C equivale a {temp_fahrenheit}°F")
```
